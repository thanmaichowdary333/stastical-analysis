 Interpreting Regression Model Results: Coefficients, P-Values, and R-Squared
Import Libraries: Load necessary libraries for data handling and modeling.
Extract Coefficients: Get values showing the impact of each feature.
Extract P-Values: Get values indicating the statistical significance of each feature.
Extract R-Squared: Get the measure of how well the model fits the data.
Print Results: Display the coefficients, p-values, and R-squared value.
Interpret P-Values: Coefficients with p-values < 0.05 are considered significant.
Interpret R-Squared: Indicates the percentage of variability explained by the model.

```python
import pandas as pd
import statsmodels.api as sm
df = pd.read_csv('health.csv')
# Drop rows with missing values for the columns of interest
df_cleaned = df[['BMI', 'Age']].dropna()
# Define the independent (X) and dependent (y) variables
X = df_cleaned[['BMI']]  # Predictor variable
y = df_cleaned['Age']  # Response variable
# Add a constant (intercept) to the independent variable
X = sm.add_constant(X)
# Perform Ordinary Least Squares (OLS) regression
model = sm.OLS(y, X).fit()
# Extract coefficients, p-values, and R-squared
coefficients = model.params
p_values = model.pvalues
r_squared = model.rsquared
# Print the results in the specified format
print("Coefficients:")
print(coefficients)
print("\nP-Values:")
print(p_values)
print(f"\nR-Squared: {r_squared:.6f}")
# Interpretation of P-values
print("\nInterpretation of P-Values:")
for variable, p_value in p_values.items():
    if p_value < 0.05:
        print(f"The coefficient for '{variable}' is statistically significant (p < 0.05).")
    else:
        print(f"The coefficient for '{variable}' is not statistically significant (p >= 0.05).")
```

This code visualizes the relationship between BMI and Age from your dataset using a scatter plot and a linear regression line. The scatter plot shows individual data points, while the regression line indicates the overall trend between the variables. The plot customization enhances readability by adjusting colors and adding labels.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('health.csv')

# Define the independent variable and dependent variable
X = sm.add_constant(df['BMI'])  # Adding constant for intercept
y = df['Age']

# Fit linear regression model
model = sm.OLS(y, X).fit()

# Create a DataFrame for predictions
df['predicted'] = model.predict(X)

# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['BMI'], y=df['Age'], label='Actual Data', color='pink')
sns.lineplot(x=df['BMI'], y=df['predicted'], label='Regression Line', color='orange')

# Customize plot
plt.xlabel('BMI')
plt.ylabel('Age')
plt.title('Relationship Between BMI and Age with Regression Line')
plt.gcf().set_facecolor('black')
plt.gca().set_facecolor('black')
plt.legend()

# Display the plot
plt.show()
```
